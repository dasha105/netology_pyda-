- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
df = pd.read_csv('/Users/administrator/Documents/нетология/python для анализа данных/лекция 7. 22.08/Python_7_pandas_functions_groupby/ml-latest-small/ratings.csv')
df.head()
```

geo_data = {
